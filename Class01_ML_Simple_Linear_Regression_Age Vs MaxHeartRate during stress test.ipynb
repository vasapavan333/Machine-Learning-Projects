```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel(r"C:\Users\pavan.vasa\OneDrive - Accenture\My Learning\Target MACHINE LEARNING 6 Months\Age Vs Max Heart Rate Acheived during stress test.xlsx")
df.head()
```

```python
df.shape
```

```python
X = df[["age"]]
y = df[["maximum_heart_rate"]]

print(X.shape)
print(y.shape)
```

```python
#Plotting scatter plot to get idea on data

plt.figure(figsize=(5, 4))
plt.title("Age Vs Maximum Blood Pressure Achieved during stress test")
plt.xlabel("Age")
plt.ylabel("maximum_heart_rate")
plt.scatter(X, y, color = 'g', marker = '*')
```

```python
#Linear Regression
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
#Input the age for prediction
input_age = 70

#reshaping the age from scalar to 2d format to match with the model

input_age_reshape = np.array([[input_age]])
print(input_age_reshape.shape)
```

```python
#predicting the maximum heart rate based on age.

reg.predict(input_age_reshape)
```

```python
#Coefficient m
reg.coef_
```

```python
#Intercept b
reg.intercept_
```

```python
#y = mx + b
# x is the maixmum heart rate that we want to predict
# b is the intercept 204.289
# m is the coefficient -1.00507682
m = reg.coef_
x1 = input_age_reshape
b = reg.intercept_
y1 = m * x1 + b
print(y1)
```

```python
#Plotting line with the predicted values.

plt.figure(figsize=(5, 4))
plt.title("Age Vs Maximum Blood Pressure Achieved during stress test")
plt.xlabel("Age")
plt.ylabel("maximum_heart_rate")
plt.scatter(X, y, color = 'r', marker = '*')
plt.plot(X, reg.predict(X), color = 'blue')
```

```python
#Using model and generating results for another set of the data.

new_data = np.array([50, 25, 65, 45, 86, 14, 30, 47])
print(new_data.shape)
new_data_2d = new_data.reshape(-1,1)
print(new_data_2d.shape)
```

```python
#converting 2d array to pandas data frame 
new_predictions = pd.DataFrame(new_data_2d, columns = ["age"])  
print(new_predictions)
```

```python
#Predicting maximum heart rates for age category using trianing model
r = reg.predict(new_predictions)
new_predictions["Maximum Heart Rate Predicted"] = r
print(new_predictions.head)
```
