```python
#Importing Required Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Loading house prices history data
df = pd.read_csv(r"houseprices.csv")
df.head()
```

Data Pre-Processing managing NaN null values

```python
#Need to manage the missing values in the bedroom
import math

median_bedrooms = math.floor(df.bedrooms.median())

print(median_bedrooms)
```

```python
#We need to fill all NaA values with the median value of the bedrooms.

df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)
```

```python
#Assign updated bedrooms column to main dataset
df.head()
```

Creating Linear Regression Model - for multiple variables

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
#Looking at coefficients m1, m2, m3 etc.,
reg.coef_
```

```python
#Looking at intercept b 
reg.intercept_
```

```python
#Prediction of house prices for 3000 square feet house with 3 bedrooms and 40 years old
reg.predict([[3000, 3, 40]])
```
