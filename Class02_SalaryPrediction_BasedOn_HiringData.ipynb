Salary Recommendation based on experience, test_score, interview_score and number of years of experience

```python
#Importing required models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
#Importing dataset

df = pd.read_excel("hiring.xlsx")
df
```

Data Cleaning

```python
#Experience can be assumed as 0 & the test_score can be asumed as median
print(df["test_score(out of 10)"].mean())
print(df["test_score(out of 10)"].median())
print(df["test_score(out of 10)"].mode())
```

```python
#Since this are the test scores, we can go with the median value as it is lower than the mean for error handling

df['test_score(out of 10)'] = df ['test_score(out of 10)'].fillna(math.floor(df["test_score(out of 10)"].mean()))
```

```python
df
```

```python
df['experience'] = df['experience'].fillna("zero")
```

```python
df
```

```python
from word2number import w2n
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["experience", "test_score(out of 10)", "interview_score(out of 10)"]], df["salary"])
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg_ = linear_model.LinearRegression()
reg_.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary'])
reg_.predict([[10,10,10]])
```
